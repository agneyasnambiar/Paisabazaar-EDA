## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

three hypothetical statements clearly defined based on the dataset and EDA charts

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

 **Null Hypothesis (H₀):** There is no significant difference in Annual Income between customers with Good and Poor credit scores.

**Alternate Hypothesis (H₁):** Customers with Good credit scores have significantly higher Annual Income than those with Poor credit scores.

Business Relevance: Determines if income level is a strong predictor of creditworthiness.

Test Used: Independent t-test (since income is continuous and we're comparing two groups).

Conclusion Basis: If p-value < 0.05, reject the null hypothesis.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import ttest_ind

good_income = df[df['Credit_Score'] == 'Good']['Annual_Income']
poor_income = df[df['Credit_Score'] == 'Poor']['Annual_Income']
t_stat, p_val = ttest_ind(good_income, poor_income)
print("T-Statistic:", t_stat)
print("P-Value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

Independent t-test

##### Why did you choose the specific statistical test?

Used to compare means between two independent groups (Good vs Poor).

If p-value < 0.05, we reject the null hypothesis and accept that Good credit score customers earn significantly more.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): The mean Credit Utilization Ratio is the same across all credit score categories.

Alternate Hypothesis (H₁): At least one credit score category has a different mean Credit Utilization Ratio.

Business Relevance: Identifies if credit behavior (utilization) impacts credit ratings.

Test Used: One-Way ANOVA (since comparing a numerical feature across 3 groups: Good, Standard, Poor).

Conclusion Basis: If p-value < 0.05, significant differences exist among groups.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import f_oneway

group1 = df[df['Credit_Score'] == 'Good']['Credit_Utilization_Ratio']
group2 = df[df['Credit_Score'] == 'Standard']['Credit_Utilization_Ratio']
group3 = df[df['Credit_Score'] == 'Poor']['Credit_Utilization_Ratio']
f_stat, p_val = f_oneway(group1, group2, group3)
print("F-Statistic:", f_stat)
print("P-Value:", p_val)
````

##### Which statistical test have you done to obtain P-Value?

One-way ANOVA

##### Why did you choose the specific statistical test?

To compare means of Credit Utilization Ratio across more than 2 groups.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H₀): There is no association between Credit Score and Occupation.

Alternate Hypothesis (H₁): There is a significant association between Credit Score and Occupation.



Business Relevance: Determines if certain professions tend to be more creditworthy.

Test Used: Chi-square Test of Independence (categorical vs categorical).

Conclusion Basis: If p-value < 0.05, reject the null and conclude a relationship exists.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Credit_Score'], df['Occupation'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
````

 Used to check relationship between two categorical variables.

#### What all missing value imputation techniques have you used and why did you use those techniques?

Upon inspection using `df.isnull().sum()`, we found that the dataset does **not contain any missing values**.

However, if missing values had been present, we would have used the following imputation techniques:

- **Numerical Features:**  
  Imputed using the **median** to avoid distortion caused by outliers. For example:  
  `df['Annual_Income'].fillna(df['Annual_Income'].median(), inplace=True)`

- **Categorical Features:**  
  Imputed using the **mode** (most frequent category). For example:  
  `df['Occupation'].fillna(df['Occupation'].mode()[0], inplace=True)`

These techniques are preferred for maintaining the central tendency of data without introducing bias, especially when the data is skewed.


### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
sns.boxplot(df['Outstanding_Debt'])
plt.title('Outstanding Debt Outliers')
plt.show()

sns.boxplot(df['Num_Credit_Inquiries'])
plt.title('Credit Inquiries Outliers')
plt.show()

# Apply IQR method
for col in ['Outstanding_Debt', 'Num_Credit_Inquiries']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] > upper, upper, df[col])
    df[col] = np.where(df[col] < lower, lower, df[col])
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

I used the **Interquartile Range (IQR) method** to identify and treat outliers in the numerical features:

- **Columns Treated:**
  - `Outstanding_Debt`
  - `Num_Credit_Inquiries`

- **Why IQR Method?**
  - It is a robust technique that is not influenced by extreme values.
  - It defines outliers as values that fall below **Q1 − 1.5×IQR** or above **Q3 + 1.5×IQR**, which helps isolate anomalous data without removing large portions of the dataset.

- **Implementation:**

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

Accuracy represents the overall correctness of the model’s predictions. While high accuracy is desirable, it alone may not be enough for imbalanced datasets like credit score classification, where one category (like 'Good') dominates. Relying only on accuracy may give a false impression of performance.

Precision indicates how many of the predicted cases for a particular credit score (e.g., ‘Poor’) were actually correct. High precision in predicting 'Poor' credit scores is crucial for minimizing the rejection of eligible customers or false red flags. From a business perspective, it helps avoid wrongly penalizing good customers.

Recall tells us how many actual cases of a given category the model successfully identified. High recall for 'Poor' ensures that most risky applicants are correctly flagged. This reduces the risk of giving loans to potentially defaulting individuals, which is essential for fraud detection and financial stability.

F1-Score is the harmonic mean of precision and recall. It’s especially important when there is a need to balance both — for example, when both missing high-risk customers and falsely tagging low-risk ones can lead to significant financial implications. A high F1-score means the model handles both concerns effectively.
